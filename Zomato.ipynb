```python
numtype=['rates(out of 5)','votes','approx_cost(for two people)']

chartype=['name','online_order','book_table','listed_in(type)']
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
axes=axes.ravel()

for i,col in enumerate({'rates(out of 5)','approx_cost(for two people)'}):
    axes[i].hist(df2[col],edgecolor='black')
    axes[i].set_title(f'Frequency Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
```

```python
fig,axes=plt.subplots(2,2,figsize=(12,8))
axes=axes.ravel()

for i,col in enumerate({'online_order','book_table','listed_in(type)'}):
    sns.countplot(df,x=col,ax=axes[i],edgecolor='black',palette='BuGn')
    axes[i].set_title(f'Frequency Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
```

```python
avgrating=df2.groupby('listed_in(type)')['rates(out of 5)'].mean()
```

```python
plt.figure(figsize=(8,6))
plt.bar(avgrating.index,avgrating.values,edgecolor='black',color=['green','blue','red','orange'])
plt.xlabel('Restaurant Type')
plt.ylabel('Rating')
plt.title('Average Rating of different types of Restaurants');
```

```python
plt.figure(figsize=(8,5))
sns.countplot(df2,x='listed_in(type)',hue='book_table',palette='colorblind')
plt.xlabel('Restaurant Types')
plt.title('Availability of Table Reservation Facility in different Restaurant types');
```

```python
plt.figure(figsize=(8,5))
sns.countplot(df2,x='listed_in(type)',hue='online_order',palette='YlGn')
plt.xlabel('Restaurant Types')
plt.title('Online Order option according to Restaurant types');
```
